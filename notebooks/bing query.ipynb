```python
@lru_cache(maxsize=None)
def get_text_from_url(url):
    resp = requests.get(url)
    txt = resp.text
    soup = BeautifulSoup(txt, 'html.parser')
    
    # filtering some elements for a clearer text.
    
    # filtering tables
    for table in soup.find_all("table"):
        table.extract()
    
    # filtering subscripts, which mess up the parsing sometimes (e.g. https://en.wikipedia.org/wiki/Kissyfur)
    for sup in soup.find_all("sup"):
        sup.extract()
    
    return soup.get_text()
```

## Series (P449)

```python
def parse_series_text(text, verb_lemma, preposition='on'):
    """
    The 'broadcast_on' rule was not part of the automated generated rules, but noticed from the data
    specifically - https://en.wikipedia.org/wiki/Homeland_(TV_series)
    """
    doc = nlp(text)
    for w in doc:
        if w.lemma_ == verb_lemma:
            for c in w.children:
                if c.text == preposition:
                    for cc in c.children:
                        if cc.ent_type_ in ['DATE']:
                            continue
                        
                        # dealing with nouns
                        # based on the following serie:
                        # https://en.wikipedia.org/wiki/Cuts_(TV_series)
                        # (Cuts is an American sitcom that aired on the UPN network from February 14, 2005, to May 11, 2006, and is a spin-off of another UPN series, One on One.)
                        # ignoring the noun, and looking for the proper noun
                        if cc.pos_ in ['NOUN']:
#                             print(cc)
                            for ccc in cc.children:
#                                 print(ccc, ccc.dep_, ccc.pos_)
                                if ccc.dep_ in ['compound'] and ccc.pos_ in ['PROPN']:
                                    return ccc.text
                        return cc.text
    return None

# achieving 35.5% accuracy on the training set (1000 examples)
```

```python
filt_lines = lines[:100]
acc = 0
total = len(filt_lines)
for entry in tqdm(filt_lines):
    subj = entry['sub_label']
    obj = entry['obj_label']
    
    url = search_results("wikipedia serie {}".format(subj))
    if url is None:
        print('no url')
        continue
    
    url_text = get_text_from_url(url)
    print(subj, obj)
    for line in url_text.split('\n'):
        if line.strip() == '':
            continue
        if 'premier' in line:
            print(line)
            ans = parse_series_text(line, 'premiere', preposition='on')
            if ans is not None:
                if ans == obj:
                    acc += 1
                print('ans:', ans)
                break
        if 'air' in line:
            print(line)
            ans = parse_series_text(line, 'air', preposition='on')
            if ans is not None:
                if ans == obj:
                    acc += 1
                print('ans:', ans)
                break
        if 'broadcast' in line:
            print(line)
            ans = parse_series_text(line, 'broadcast', preposition='on')
            if ans is not None:
                if ans == obj:
                    acc += 1
                print('ans:', ans)
                break
    print()
```

```python
def parse_speaks(text):
    doc = nlp(text)
    for w in doc:
        if w.lemma_ in ['speak']:
            for c in w.children:
                if c.dep_ == 'dobj':
                    return c.text
    return None
```

```python
filt_lines = lines[:1000]
acc = 0
total = len(filt_lines)
for entry in tqdm(filt_lines):
    subj = entry['sub_label']
    obj = entry['obj_label']
    
    url = search_results("wikipedia location {}".format(subj))
    if url is None:
        print('no url')
        continue
    
    url_text = get_text_from_url(url)
    url_text = re.sub(r' \([^)]*\)', '', url_text)
    url_text = re.sub(r'\([^)]*\) ', '', url_text)
    url_text = re.sub(r'\([^)]*\)', '', url_text)
#     url_text = url_text.replace(r" \(.*\)", "")
    
    print(subj, obj)
    for line in url_text.split('\n'):
#         print(line)
        if line.strip() == '':
            continue
        if 'speak' in line:
            print(line)
            ans = parse_speaks(line)
            if ans is not None:
                if ans == obj:
                    acc += 1
                print('ans:', ans)
                break
    print()
```

```python
def parse_subclass_of(title, text):
    doc = nlp(text)
    for w in doc:
        if w.text == title:
            if w.dep_ == 'compound':
                if w.head.pos_ == 'NOUN':
                    return w.head.text
    return None
```

```python
def parse_is_a(text):
    doc = nlp(text)
    for w in doc:
        if w.text == 'is':
            for c in w.children:
                if c.dep_ == 'attr':
                    if c.pos_ == 'NOUN':
                        return c.text
    return None
```

```python
filt_lines = lines[:1000]
acc = 0
total = len(filt_lines)
for entry in tqdm(filt_lines):
    subj = entry['sub_label']
    obj = entry['obj_label']
    
    url = search_results("wikipedia {}".format(subj))
    if url is None:
        print('no url')
        continue
    title = url.split('/')[-1]
    
    url_text = get_text_from_url(url)
    
    print('new rel', subj, obj)
    for line in url_text.split('\n'):
#         print(line)
        if line.strip() == '':
            continue
        if title in line:
            print(line)
            ans = parse_subclass_of(title, line)
            if ans is not None:
                if ans == obj:
                    acc += 1
                print('ans:', ans)
                break
        if 'is a' in line:
            ans = parse_is_a(line)
            if ans is not None:
                if ans == obj:
                    acc += 1
                print('ans:', ans)
                break
    print()
```
