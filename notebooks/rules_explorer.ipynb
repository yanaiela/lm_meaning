## P136

```python
data = read_file('../data/TREx_train/{}.jsonl'.format('P136'))
```

```python

class P136(RuleMatcher):

    def __init__(self):
        search_query = "wikipedia genre {}"
        super().__init__(search_query)

    def match_rules(self, line, params={}):
        
        if 'infobox' in params:
            ans = self.parse_infobox(params['infobox'])
            if ans is not None:
                return ans
            
        if 'is a' in line:
            ans = self.parse_is_a(line)
            if ans is not None:
                return ans

        return None

    def parse_is_a(self, text):
        doc = self.nlp(text)
        for w in doc:
            if w.lemma_ in ['be']:
                for c in w.children:
                    if c.dep_ == 'attr':# and c.pos_ == 'NOUN':
                        return c.text
        return None

    def parse_infobox(self, infobox):
#         print('in info box', infobox)
        # Genres in the infobox: https://en.wikipedia.org/wiki/En_Nuestros_Corazones
        for entry in infobox:
            if entry['key'] in ['Genres', 'Genre']:
                return entry['values'][0].lower()
        return None
```
